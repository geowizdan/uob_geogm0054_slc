```python
import numpy as np
import matplotlib.pyplot as plt
from calendar import month_name
```

```python
year, month, tmax, tmin = np.loadtxt('rossonwyedata.txt',skiprows=7,usecols=(0,1,2,3),unpack=True)
```

```python
month_nums = range(1,13) # 1 to 12, inclusive
month_names = [month_name[mm] + f' ({mm})'  for mm in month_nums]

#create a list of 1D arrays, 1 for each month
tmax_by_month = [tmax[month == mm] for mm in month_nums]

#HINT reverse the month order to place January at the top 
tmax_by_month.reverse() 
month_names.reverse()

#boxplot produces one box-and whisker for each 1D array in a list
fig, ax = plt.subplots(1,1)
__ = ax.boxplot(tmax_by_month, vert=False, tick_labels = month_names)
# HINT add the keyword argument labels=month_names (Python < 3.9) or tick_labels=month_names (Python >= 3.9)
__ = ax.set_xlabel(r'Temperature ($^\circ$C)')
__ = ax.set_ylabel('Month')
__ = ax.set_title(f'Ross-on-Wye maximum temperatures, {int(np.min(year))} to {int(np.max(year))}')

   
```

```python
# Boxplot produces one box-and-whisker for each 1D array in a list
# Add color to the boxplot
boxprops = dict(color='blue', linewidth=0.5)
medianprops = dict(color='red', linewidth=0.5)
whiskerprops = dict(color='green', linewidth=0.5)
capprops = dict(color='purple', linewidth=0.5)



fig, ax = plt.subplots(1, 1, figsize=(10, 6))
ax.boxplot(tmax_by_month, vert=False, tick_labels=month_names, 
          boxprops=boxprops, medianprops=medianprops,
           whiskerprops=whiskerprops, capprops=capprops)

# Set labels and title
ax.set_xlabel(r'Temperature ($^\circ$C)')
ax.set_ylabel('Month')
ax.set_title(f'Ross-on-Wye maximum temperatures, {int(np.min(year))} to {int(np.max(year))}')

# Improve the plot aesthetics
ax.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()

```

```python
# Define colors for each month
colors = plt.cm.Dark2(np.linspace(0, 1, len(month_nums)))
medianprops = dict(color='red', linewidth=0.8)
whiskerprops = dict(color ='black', linewidth =0.5)
# Boxplot produces one box-and-whisker for each 1D array in a list
fig, ax = plt.subplots(1, 1, figsize=(12, 8))

# Create the boxplot with patch_artist=True to fill the boxes with color
boxplot = ax.boxplot(tmax_by_month, vert=False, patch_artist=True, tick_labels=month_names, medianprops=medianprops,
           whiskerprops=whiskerprops)

# Fill the boxes with different colors
for patch, color in zip(boxplot['boxes'], colors):
    patch.set_facecolor(color)

# Set labels and title
ax.set_xlabel(r'Temperature ($^\circ$C)', fontsize=14)
ax.set_ylabel('Month', fontsize=14)
ax.set_title(f'Ross-on-Wye Maximum Temperatures\n{int(np.min(year))} to {int(np.max(year))}', fontsize=14)

# Improve the plot aesthetics
ax.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```
